```python
import dognet
from dognet import training
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import math
```

```python
focal = training.focal_loss(alpha = .5,gamma = 2)
bce = nn.BCELoss()

tar = torch.empty(1,1)
tar[0][0] = 0
pred = torch.empty(1,1)

bceloss = []
focalloss = []

r = np.arange(0.1,1,.1)
for i in r:
    pred[0][0] = i

    bceloss.append(bce(pred,tar))
    focalloss.append(focal.loss(pred,tar))

plt.plot(r,bceloss)
plt.plot(r,focalloss)
plt.legend(["bce","focal"])
```

```python
#distance
focal = training.focal_loss(alpha = .5,gamma = 0)
bce = nn.BCELoss()
L2 = nn.MSELoss()

cpy = torch.empty(800,800)

for i in range(385,416):
    for j in range(385,416):
        cpy[i][j]=1
        
sdlosses = []
fclosses = []
bcelosses = []
L2losses = []
bceq = []

dists = np.arange(0,30,1)

for d in dists:
    img = torch.empty(800,800)
    for i in range(385-d,416-d):
        for j in range(385-d,416-d):
            img[i][j]=1
            
    #sdlosses.append(np.asarray(training.soft_dice_loss(cpy, img).item()))
    #fclosses.append(np.asarray(focal.loss(cpy, img).item()))
    #bcelosses.append(np.asarray(bce(cpy, img).item()))
    #L2losses.append(np.asarray(L2(cpy, img).item()))
    sdlosses.append(training.soft_dice_loss(cpy, img))
    L2losses.append(L2(cpy, img))
    fclosses.append(focal.loss(cpy, img))
    bcelosses.append(bce(cpy, img))
```

```python
plt.plot(sdlosses)
plt.plot(fclosses)
plt.plot(bcelosses)
plt.plot(L2losses)
plt.legend(["sd", "focal", "bce", "L2"])
```

```python
cpy = torch.empty(800,800)

for i in range(385,416):
    for j in range(385,416):
        cpy[i][j]=1
        
sdlosses = []
fclosses = []
bcelosses = []
L2losses = []

dists = np.arange(0,30,1)
bright = np.arange(0,1.1,.1)

for d in dists:
    for b in bright:
        img = torch.empty(800,800)
        for i in range(385-d,416-d):
            for j in range(385-d,416-d):
                img[i][j]=b
            
        sdlosses.append(np.asarray(training.soft_dice_loss(cpy, img).item()))
        fclosses.append(np.asarray(focal.loss(y_pred = cpy, y_tar = img).item()))
        bcelosses.append(np.asarray(bce(cpy, img).item()))
        L2losses.append(np.asarray(L2(cpy, img).item()))
```

```python
from mpl_toolkits import mplot3d
%matplotlib inline

fig = plt.figure()
ax = plt.axes(projection='3d')

X,Y = np.meshgrid(bright,dists)
Z=np.reshape(sdlosses,(30,11))
Z1=np.reshape(fclosses,(30,11))
Z2=np.reshape(bcelosses,(30,11))
Z3=np.reshape(L2losses,(30,11))


#ax.contour3D(X,Y,Z,100,cmap = 'binary')
ax.contour3D(X,Y,Z1,100,cmap = 'pink')
ax.contour3D(X,Y,Z2,100,cmap = 'autumn')
#ax.contour3D(X,Y,Z3,100,cmap = 'bone')

ax.set_xlabel("bright")
ax.set_ylabel("dist")
ax.set_zlabel("error")
```
